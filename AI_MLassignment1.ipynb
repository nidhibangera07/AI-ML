```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [],
    'E': [],
    'F': [],
    'G': []
}

visited = set()

def dfs(node):
    if node not in visited:
        visited.add(node)
        print(node, end=" ")

        for neighbor in graph[node]:
            dfs(neighbor)

print("DFS Traversal from 'A':")
dfs('A')
```
